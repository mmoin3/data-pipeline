```python
import os, sys
from pathlib import Path
import pandas as pd
sys.path.append(str(Path("..").resolve()))
from config import *
from src.services.statestreet_mft_client import StateStreetMFTClient
```

```python
with StateStreetMFTClient() as client:
    files = client.list_files("/ETFGlobalHarvest/fromSSC")
    downloaded_files = []
    
    for file_info in files:
        if not file_info.get("directory"):
            filename = file_info.get("filename")
            remote_path = f"/ETFGlobalHarvest/fromSSC/{filename}"
            file_path = client.download(remote_path, skip_existing=True)
            downloaded_files.append(file_path)
```
